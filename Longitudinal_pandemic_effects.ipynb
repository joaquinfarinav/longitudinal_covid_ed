```python
def create_dataframe(dictionary, length):
    df = pd.DataFrame()
    for key, data in dictionary.items():
        labels = data.label.tolist()
        varnames = data.varname.tolist()

        # Calculate the number of missing entries
        missing_entries = length - len(labels)

        # Add missing entries (NaNs) to the 'labels' and 'varnames' lists
        labels.extend([np.nan] * missing_entries)
        varnames.extend([np.nan] * missing_entries)

        # Add 'label' and 'varname' lists as new columns in df DataFrame
        if '_fa' in key:
            key = key[:7] + '_fa'
        elif '_fb' in key:
            key = key[:7] + '_fb'
        elif '_lect' in key:
            key = key[:7] + '_lect'
        elif '_mat' in key:
            key = key[:7] + '_mat'
        elif '_soc' in key:
             key = key[:7] + '_soc'
        elif '_nat' in key:
             key = key[:7] + '_nat'
        else:
            key = key[:7]

        df[f'{key}_question_text'] = labels
        df[f'{key}_quesiton_code'] = varnames

    return df.dropna(how='all')

df_cpad = create_dataframe(cpad_dict, 200)
df_cest = create_dataframe(cest_dict, 240)
df_cprof = create_dataframe(cprof_dict, 240)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def find_similar_questions(df, columns, similarity_threshold=0.7, lower_similarity_threshold=0.5):
    # Create a copy of the DataFrame with only the desired columns and drop rows where all columns are NaN
    df_subset = df[columns].dropna(how='all')

    # Fill any remaining NaN values with blank space
    df_subset = df_subset.fillna(' ')

    # Combine the text from all columns and convert to a TF-IDF matrix
    combined_text = df_subset[columns[0]].tolist()
    for column in columns[1:]:
        combined_text += df_subset[column].tolist()

    vectorizer = TfidfVectorizer().fit_transform(combined_text)

    # Split the combined vectorized data back into separate groups for each column
    vectors = [vectorizer[sum(len(df_subset[column].tolist()) for column in columns[:i]):sum(len(df_subset[column].tolist()) for column in columns[:i+1])] for i in range(len(columns))]

    # To avoid duplicity, we'll keep track of seen question pairs
    seen_pairs = set()
    similar_questions_list = []  # we'll store each row as a dict in this list, and then convert this list to a DataFrame

    # Compute the cosine similarity between each pair of columns
    for i in range(len(columns)):
        for j in range(i+1, len(columns)):
            similarities = cosine_similarity(vectors[i], vectors[j])

            # Consider questions as similar if cosine similarity is greater than the similarity_threshold
            similar_questions_indices = np.where(similarities > similarity_threshold)

            for idx in range(len(similar_questions_indices[0])):
                row_idx_i = similar_questions_indices[0][idx]
                row_idx_j = similar_questions_indices[1][idx]

                # Check if we've already seen this pair of questions
                question_i = df_subset[columns[i]].iloc[row_idx_i]
                question_j = df_subset[columns[j]].iloc[row_idx_j]
                if (question_i, question_j) in seen_pairs or (question_j, question_i) in seen_pairs:
                    continue
                
                similar_questions = {columns[i]: question_i, 
                                     columns[j]: question_j}

                for k in range(len(columns)):
                    if k != i and k != j:
                        similarities_to_k = cosine_similarity(vectors[i][row_idx_i], vectors[k])
                        max_similarity_idx = np.argmax(similarities_to_k)

                        if similarities_to_k[0][max_similarity_idx] > lower_similarity_threshold:
                            similar_questions[columns[k]] = df_subset[columns[k]].iloc[max_similarity_idx]
                        else:
                            similar_questions[columns[k]] = None

                similar_questions_list.append(similar_questions)
                seen_pairs.add((question_i, question_j))

    similar_questions_df = pd.DataFrame(similar_questions_list)

    return similar_questions_df
```

# Match T6 = ['2011_4b', '2013_6b', '2017_2m']


```python
for df in [df_cpad,df_cest,df_cprof]:
    cols = [x for x in df.columns if (x[:7] in ['2011_4b', '2013_6b','2017_2m'] ) and ('_question_text' in x) ]
    print('\n',cols)
    df_results = find_similar_questions(df, cols , 0.7, 0.5)
    
    for ind, row in df_results.dropna().iterrows():
        print(row.values, '\n')
```
